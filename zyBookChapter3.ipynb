# Heart Disease

```python
data = {'Diagnosis': ['Heart disease', 'No heart disease', 'Total'],
        'Low HDL': [27, 63, 90],
        'Healthy HDL': [12, 398, 410],
        'Total': [39, 461, 500]}

heart = pd.DataFrame(data)
heart.head()
```

```python
total = heart['Total'].iloc[-1]  # Total count from the last row
heart_prob = heart.iloc[:-1, 1:].div(heart.iloc[:-1, -1], axis=0)  # Divide each cell by row total, excluding the 'Total' row

# Add back Diagnosis column
heart_prob.insert(0, 'Diagnosis', heart['Diagnosis'][:-1])

# Print the conditional probabilities table
print(heart_prob)
```

```python
p_heartDisease = 39/500
p_low_givenHeartDisease = 27/39
p_healthy_givenHeartDisease = 12/39
p_low_givenNoHeartDisease = 63/461
p_noHeartDisease = 1-p_heartDisease

like_heartDisease_given_low = heart_prob['Low HDL'].iloc[0]*p_heartDisease
like_noHeartDisease_given_low = heart_prob['Low HDL'].iloc[1]*p_noHeartDisease

print(f"{p_heartDisease=}")
print(f"{p_low_givenHeartDisease=}")
print(f"P_heartDisease_given_low= {like_heartDisease_given_low/(like_heartDisease_given_low+like_noHeartDisease_given_low)=}")
print(f"P_noHeartDisease_given_low= {like_noHeartDisease_given_low/(like_heartDisease_given_low+like_noHeartDisease_given_low)=}")


```
