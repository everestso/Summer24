```python
import numpy as np
import pandas as pd
```

```python
import requests

# Replace with the actual shareable link
bitly_link = 'https://bit.ly/zyBookDiamonds'
file_name = 'diamonds.csv'

response = requests.head(bitly_link, allow_redirects=True)
file_url = response.url

# Extract the file ID from the URL
file_id = file_url.split('/')[-2]

# Construct the download URL
download_url = f'https://drive.google.com/uc?id={file_id}'

# Download the file
response = requests.get(download_url)

# Save the file temporarily
with open(file_name, 'wb') as f:
    f.write(response.content)

# Read the downloaded file into a pandas DataFrame
df = pd.read_csv(file_name)

# Display the first few rows of the DataFrame
df.head()
```

```python
diamond_sample = df.sample(1000, random_state=12)

# Create a dataframe X containing all the features except cut, color, clarity, and price
X = diamond_sample.drop(['cut', 'color', 'clarity', 'price'], axis=1)
print(f"{X.columns=}, {X.shape=}")
print(f"{X[:3]}")


# Create a dataframe y containing the feature price
y = diamond_sample[['price']]
print(f"{y.columns}")
print(f"{y.iloc[:3]=}")
```
